```python
import pandas as pd
```

```python
df=pd.read_csv('../data/bakery.csv')
df
```

```python
selected_columns = ['data-product-id','data-product-name','data-product-type']
print(df[selected_columns])
```

```python
if df['product-name'].equals (df['data-product-name']):
    print(" The columns are identical ")
    df = df.drop(columns=['data-product-name'])

else:
    print("The columns are not identical. Dropping the Columns")
```

```python
df = df.rename(columns= {
                'data-product-id':'product-id',
                'data-product-name':'pro_name',
                'data-product-type':'pro_type',
                'data-product-on-offer':'offer',
                'data-product-index':'index',
                'image-url':'url',
                'product-page':'page',
                'product-name':'name',
                'product-size':'size',
                'item-price':'price',
                'price-per-unit':'unit',
                'offer-description':'offer_desc',
                'category':'catgs'}
                )
df
```

```python
item_price_list =df['price'].tolist()

new_list = []

for element in item_price_list:
    if '£' in element:
        new_element = element.replace('£', '')
    elif 'p' in element:
        new_element = element.replace('p','')
        new_element = '0.' + new_element
    else:
        new_element = element

    new_list.append(new_element)

df['price'] = new_list
df['price'] = df['price'].astype(float)
```
